```python
import matplotlib.pyplot as plt
import numpy as np
import math
import random
def get_signal(length,num_layers=5):
    random.seed(1)
    array = np.zeros(length)
    for i in range(0,length):
        for j in range(1,num_layers+1):
            array[i]+=math.sin(float(i)/(float(j)*float(length)*0.001))+random.random()
    return array
plt.plot(get_signal(10000))
plt.show()
    
```
